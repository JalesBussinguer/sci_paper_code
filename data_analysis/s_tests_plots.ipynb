```python
dprvi_f = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/florestalDpRVI_reindex.csv')
dprvi_s = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/savanicaDpRVI_reindex.csv')
dprvi_c = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/campestreDpRVI_reindex.csv')

prvi_f = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/florestalPRVI_reindex.csv')
prvi_s = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/savanicaPRVI_reindex.csv')
prvi_c = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/campestrePRVI_reindex.csv')

dpsvi_f = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/florestalDPSVI_reindex.csv')
dpsvi_s = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/savanicaDPSVI_reindex.csv')
dpsvi_c = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/campestreDPSVI_reindex.csv')

dpsvim_f = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/florestalDPSVIm_reindex.csv')
dpsvim_s = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/savanicaDPSVIm_reindex.csv')
dpsvim_c = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/campestreDPSVIm_reindex.csv')

rvi_f = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/florestalRVI_reindex.csv')
rvi_s = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/savanicaRVI_reindex.csv')
rvi_c = pd.read_csv('D:/thesis_data/VEG_INDICES/stats/campestreRVI_reindex.csv')

ff_rates = pd.read_csv('D:/thesis_data/VEG_INDICES/rates/FF_rates.csv')
fs_rates = pd.read_csv('D:/thesis_data/VEG_INDICES/rates/FS_rates.csv')
fc_rates = pd.read_csv('D:/thesis_data/VEG_INDICES/rates/FC_rates.csv')
```

```python
dprvi_f['data'] = pd.to_datetime(dprvi_f['date'], format='%Y/%m/%d')

dprvi_f['month'] = dprvi_f['data'].dt.month

dprvi_f.loc[dprvi_f['month']==1,'mes'] = 'Jan'
dprvi_f.loc[dprvi_f['month']==2,'mes'] = 'Feb'
dprvi_f.loc[dprvi_f['month']==3,'mes'] = 'Mar'
dprvi_f.loc[dprvi_f['month']==4,'mes'] = 'Apr'
dprvi_f.loc[dprvi_f['month']==5,'mes'] = 'May'
dprvi_f.loc[dprvi_f['month']==6,'mes'] = 'Jun'
dprvi_f.loc[dprvi_f['month']==7,'mes'] = 'Jul'
dprvi_f.loc[dprvi_f['month']==8,'mes'] = 'Aug'
dprvi_f.loc[dprvi_f['month']==9,'mes'] = 'Sep'
dprvi_f.loc[dprvi_f['month']==10,'mes'] = 'Oct'
dprvi_f.loc[dprvi_f['month']==11,'mes'] = 'Nov'
dprvi_f.loc[dprvi_f['month']==12,'mes'] = 'Dec'
```

```python
# Temporal profiles

fig, ax = plt.subplots(3, 1, figsize=(20, 15))

ax[0].plot(ff_rates.index, ff_rates['dprvi'], marker='o', label='DpRVI')
ax[0].plot(ff_rates.index, ff_rates['prvi'], marker='^', label='PRVI')
ax[0].plot(ff_rates.index, ff_rates['dpsvi'], marker='s', label='DPSVI')
ax[0].plot(ff_rates.index, ff_rates['dpsvim'], marker='d',label='DPSVIm')
ax[0].plot(ff_rates.index, ff_rates['rvi'], marker='*', label='RVI')
ax[0].set_xticks(ticks=dprvi_f.index)
ax[0].tick_params(axis='y', length=5, width=1.5, labelsize=25)
ax[0].set_xlim([-1, 149])
ax[0].set_ylabel('FF', fontsize=30)
ax[0].locator_params(axis='x', nbins=60)
ax[0].locator_params(axis='y', nbins=8)
ax[0].legend()

ax[1].plot(fs_rates.index, fs_rates['dprvi'], marker='o', label='DpRVI')
ax[1].plot(fs_rates.index, fs_rates['prvi'], marker='^', label='PRVI')
ax[1].plot(fs_rates.index, fs_rates['dpsvi'], marker='s', label='DPSVI')
ax[1].plot(fs_rates.index, fs_rates['dpsvim'], marker='d',label='DPSVIm')
ax[1].plot(fs_rates.index, fs_rates['rvi'], marker='*', label='RVI')
ax[1].tick_params(axis='y', length=5, width=1.5, labelsize=25)
ax[1].set_xticks(ticks=dprvi_f.index)
ax[1].set_xlim([-1, 149])
ax[1].set_ylabel('FS', fontsize=30)
ax[1].locator_params(axis='x', nbins=60)
ax[1].locator_params(axis='y', nbins=8)
ax[1].legend()


ax[2].plot(fc_rates.index, fc_rates['dprvi'], marker='o', label='DpRVI')
ax[2].plot(fc_rates.index, fc_rates['prvi'], marker='^', label='PRVI')
ax[2].plot(fc_rates.index, fc_rates['dpsvi'], marker='s', label='DPSVI')
ax[2].plot(fc_rates.index, fc_rates['dpsvim'], marker='d',label='DPSVIm')
ax[2].plot(fc_rates.index, fc_rates['rvi'], marker='*', label='RVI')
ax[2].tick_params(axis='y', length=5, width=1.5, labelsize=25)
ax[2].set_xticks(ticks=dprvi_f.index)
ax[2].set_xlim([-1, 149])
ax[2].set_ylabel('FS', fontsize=30)
ax[2].locator_params(axis='x', nbins=60)
ax[2].locator_params(axis='y', nbins=8)
ax[2].legend()

plt.tight_layout()
```
