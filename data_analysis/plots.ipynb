```python
# Imports
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt

import rasterio as rst
from rasterio.mask import mask
from rasterio.plot import show_hist
```

```python
area = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_florestal_32723.GEOJSON')
```

```python
geometry = [geom for geom in area.geometry]
```

```python
with rst.open('D:/thesis_data/preprocessed/GRD/S1_Orb_Cal_Spk_Ter_Sub_20170112T084450.tif') as raster:
    out_image, out_transform = mask(raster, geometry, crop=True, nodata=np.nan)
```

```python
#fig = plt.figure(figsize=(20,10))
(n, bins) = np.histogram(out_image[1], bins=50)
plt.plot(n)
```

```python
f, ax = plt.subplots(figsize=(20,10))
plt.hist(out_image[0], bins=10)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
hist = show_hist(source=out_image[0], bins=50, title='Histogram of all bands', histtype='stepfilled', alpha=0.5)
fig.legend(hist)
```
