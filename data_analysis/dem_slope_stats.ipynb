```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dem_pnb = pd.read_csv('D:/thesis_data/TOPOGRAPHY/dem_pnb_samples.csv')
dem_pnb_valid = dem_pnb.loc[(dem_pnb["dem"] != -1)]
```

```python
dem_ff = pd.read_csv('D:/thesis_data/TOPOGRAPHY/dem_ff_samples.csv')
dem_ff_valid = dem_ff.loc[(dem_ff["dem"] != -1)]
```

```python
dem_fs = pd.read_csv('D:/thesis_data/TOPOGRAPHY/dem_fs_samples.csv')
dem_fs_valid = dem_fs.loc[(dem_fs["dem"] != -1)]
```

```python
dem_fc = pd.read_csv('D:/thesis_data/TOPOGRAPHY/dem_fc_samples.csv')
dem_fc_valid = dem_fc.loc[(dem_fc["dem"] != -1)]
```

```python
red_circle = dict(markerfacecolor='red', marker='o', markeredgecolor='white')

fig, ax = plt.subplots(1, 4, figsize=(15,10))

ax[0].boxplot(dem_pnb_valid, flierprops=red_circle)
ax[0].set_title('PNB', fontsize=20, fontweight='bold')
ax[0].tick_params(axis='y', labelsize=14)
ax[0].set_ylim([800, 1400])

ax[1].boxplot(dem_ff_valid, flierprops=red_circle)
ax[1].set_title('FF', fontsize=20, fontweight='bold')
ax[1].tick_params(axis='y', labelsize=14)
ax[1].set_ylim([800, 1400])

ax[2].boxplot(dem_fs_valid, flierprops=red_circle)
ax[2].set_title('FS', fontsize=20, fontweight='bold')
ax[2].tick_params(axis='y', labelsize=14)
ax[2].set_ylim([800, 1400])

ax[3].boxplot(dem_fc_valid, flierprops=red_circle)
ax[3].set_title('FC', fontsize=20, fontweight='bold')
ax[3].tick_params(axis='y', labelsize=14)
ax[3].set_ylim([800, 1400])

plt.tight_layout()
```

```python
slope_pnb = pd.read_csv('D:/thesis_data/TOPOGRAPHY/slope_pnb_samples.csv')
slope_pnb_valid = slope_pnb.loc[(slope_pnb['slope'] != -1)]
```

```python
slope_ff = pd.read_csv('D:/thesis_data/TOPOGRAPHY/slope_ff_samples.csv')
slope_ff_valid = slope_ff.loc[(slope_ff['slope'] != -1)]
```

```python
slope_fs = pd.read_csv('D:/thesis_data/TOPOGRAPHY/slope_fs_samples.csv')
slope_fs_valid = slope_fs.loc[(slope_fs['slope'] != -1)]
```

```python
slope_fc = pd.read_csv('D:/thesis_data/TOPOGRAPHY/slope_fc_samples.csv')
slope_fc_valid = slope_fc.loc[(slope_fc['slope'] != -1)]
```

```python
red_circle = dict(markerfacecolor='red', marker='o', markeredgecolor='white')

median_patch = dict(color='red')

pink_patch = dict(facecolor='pink', color='k')
blue_patch = dict(facecolor='lightblue', color='k')
green_patch = dict(facecolor='lightgreen', color='k')
yellow_patch = dict(facecolor='lightyellow', color='k')

fig, ax = plt.subplots(1, 4, figsize=(18,12))

ax[0].boxplot(slope_pnb_valid, patch_artist=True, flierprops=red_circle, boxprops=yellow_patch, medianprops=median_patch)
ax[0].tick_params(axis='y', labelsize=14)
ax[0].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
ax[0].set_ylim([-1, 32])
ax[0].set_ylabel('Slope [degrees]', fontsize=20)
ax[0].set_xlabel('Study Site', fontdict={'fontsize': 20})

ax[1].boxplot(slope_ff_valid, patch_artist=True, flierprops=red_circle, boxprops=pink_patch, medianprops=median_patch)
ax[1].tick_params(axis='y', labelsize=14)
ax[1].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
ax[1].set_ylim([-1, 32])
ax[1].set_xlabel('Gallery Forests', fontdict={'fontsize': 20})

ax[2].boxplot(slope_fs_valid, patch_artist=True, flierprops=red_circle, boxprops=green_patch, medianprops=median_patch)
ax[2].tick_params(axis='y', labelsize=14)
ax[2].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
ax[2].set_ylim([-1, 32])
ax[2].set_xlabel('Shrublands', fontdict={'fontsize': 20})

ax[3].boxplot(slope_fc_valid, patch_artist=True, flierprops=red_circle, boxprops=blue_patch, medianprops=median_patch)
ax[3].tick_params(axis='y', labelsize=14)
ax[3].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
ax[3].set_ylim([-1, 32])
ax[3].set_xlabel('Grasslands', fontdict={'fontsize': 20})

plt.tight_layout()

plt.savefig('C:/Users/jales/Desktop/slopes_boxplot.svg', dpi=None, facecolor='w', edgecolor='w',
        orientation='landscape', format='svg', transparent=False)
```
