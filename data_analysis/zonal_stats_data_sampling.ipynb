```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import os

from rasterstats import zonal_stats
```

```python
form_florestal = gpd.read_file('D:/thesis_data/ROI/classes/form_florestal_10m_32723.GEOJSON')
```

```python
indices = os.listdir('D:/thesis_data/VEG_INDICES/2021')
```

```python
df_stats = pd.DataFrame()

for index, image in enumerate(indices):

    if image.endswith('.tif'):

        date = indices[index].split('.')[0]

        florestal = zonal_stats(form_florestal, 'D:/thesis_data/VEG_INDICES/2021/' + str(image), band=1, nodata=np.nan, stats=['median', 'mean', 'std'])

        stats = pd.DataFrame(florestal)
        stats['date'] = date
        df_stats = pd.concat([stats, df_stats], ignore_index=True)

        print(f'data of {date} collected!')
```

```python
df_stats.head(30)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(df_stats.date, df_stats['median'], 'g', marker='o')

plt.xticks(rotation=45)
ax.set_xlabel('Date')
```
