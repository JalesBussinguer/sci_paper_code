```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ff_VH = pd.read_csv('E:/SBSR_data/CSV/form_florestais_VH_amostras.csv', sep=',')
ff_VV = pd.read_csv('E:/SBSR_data/CSV/form_florestais_VV_amostras.csv', sep=',')

fs_VH = pd.read_csv('E:/SBSR_data/CSV/form_savanica_VH_amostras.csv', sep=',')
fs_VV = pd.read_csv('E:/SBSR_data/CSV/form_savanica_VV_amostras.csv', sep=',')

fc_VH = pd.read_csv('E:/SBSR_data/CSV/form_campestre_VH_amostras.csv', sep=',')
fc_VV = pd.read_csv('E:/SBSR_data/CSV/form_campestre_VV_amostras.csv', sep=',')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(ff_VH.date, ff_VH['mediana'], '#71C42B', marker='o', label='Formação Florestal')
ax.plot(ff_VV.date, ff_VV['mediana'], '#71C42B', marker='o')

ax.plot(fs_VH.date, fs_VH['mediana'], '#ff9801', marker='o', label='Formação Savânica')
ax.plot(fs_VV.date, fs_VV['mediana'], '#ff9801', marker='o')

ax.plot(fc_VH.date, fc_VH['mediana'], '#cbd90f', marker='o', label='Formação Campestre')
ax.plot(fc_VV.date, fc_VV['mediana'], '#cbd90f', marker='o')

plt.xticks(rotation=90)
#plt.title('Formação Florestal')

#plt.ylim(-18.0, -12.0)
plt.legend()
#ax.set_xlabel('Data')
ax.set_ylabel('Mediana [dB]')

plt.savefig('C:/Users/jales/Desktop/SBSR_figures/serie_temporal2.svg', dpi=300, facecolor='w', edgecolor='w',
        orientation='landscape', format='svg', transparent=False)
```
