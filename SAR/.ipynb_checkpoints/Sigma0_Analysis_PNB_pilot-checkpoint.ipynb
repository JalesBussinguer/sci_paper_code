## Sigma0 Analysis - PNB

Pré-processamento: Orbit File -> Calibration -> Speckle Filtering -> Terrain Correction -> Spatial Subset

```python
# Importações
# Bibliotecas básicas
import rasterio as rst
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

# Rasterio
import rasterio as rst
from rasterio.mask import mask
from rasterio.plot import show
# Raster stats
from rasterstats import zonal_stats
from scipy.stats import iqr
# Processamento de dados
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
# Áreas de estudo no Parque Nacional de Brasília
PNB = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/PNB_32723.GEOJSON')
form_florestal = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_florestal_32723.GEOJSON')
form_savanica = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_savanica_32723.GEOJSON')
form_campestre = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_campestre_32723.GEOJSON')
agua = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/agua_32723.GEOJSON')

# Importação da imagem preprocessada
with rst.open('E:/SBSR_data/TIFF/S1A_IW_GRDH_1SDV_20190911_Orb_Cal_Spk_TC.tif', 'r') as image:
    VH = image.read(1)
    VV = image.read(2)
```

```python
VH_flat = VH.flatten()
VV_flat = VV.flatten()
img_data = pd.DataFrame({'VH': VH_flat,'VV': VV_flat})
```

```python
fig = plt.figure()
ax = fig.add_subplot()

x_VV = img_data['VV']
y_VH = img_data['VH']

ax.scatter(x_VV, y_VH)

ax.set_xlabel('Sigma0_VV')
ax.set_ylabel('Sigma0_VH')

plt.show()
```
