```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ff_fs = pd.read_csv('D:/thesis_data/VEG_INDICES/ovl/FF_FS_OVL_500_bins.csv')
ff_fc = pd.read_csv('D:/thesis_data/VEG_INDICES/ovl/FF_FC_OVL_500_bins.csv')
fs_fc = pd.read_csv('D:/thesis_data/VEG_INDICES/ovl/FS_FC_OVL_500_bins.csv')
```

```python
dprvi_f = pd.read_csv('D:/thesis_data/VEG_INDICES/dprvi_parameters/stats/florestal_DpRVI_reindex.csv')
```

```python
def month_season_detector(df):

    df['data'] = pd.to_datetime(df['date'], format='%Y/%m/%d')

    df['month'] = df['data'].dt.month

    df.loc[df['month']==1,'mes'] = 'Jan'
    df.loc[df['month']==2,'mes'] = 'Feb'
    df.loc[df['month']==3,'mes'] = 'Mar'
    df.loc[df['month']==4,'mes'] = 'Apr'
    df.loc[df['month']==5,'mes'] = 'May'
    df.loc[df['month']==6,'mes'] = 'Jun'
    df.loc[df['month']==7,'mes'] = 'Jul'
    df.loc[df['month']==8,'mes'] = 'Aug'
    df.loc[df['month']==9,'mes'] = 'Sep'
    df.loc[df['month']==10,'mes'] = 'Oct'
    df.loc[df['month']==11,'mes'] = 'Nov'
    df.loc[df['month']==12,'mes'] = 'Dec'

    df.loc[df['month']==1,'season'] = 'wet'
    df.loc[df['month']==2,'season'] = 'wet'
    df.loc[df['month']==3,'season'] = 'wet'
    df.loc[df['month']==4,'season'] = 'wet'
    df.loc[df['month']==5,'season'] = 'dry'
    df.loc[df['month']==6,'season'] = 'dry'
    df.loc[df['month']==7,'season'] = 'dry'
    df.loc[df['month']==8,'season'] = 'dry'
    df.loc[df['month']==9,'season'] = 'dry'
    df.loc[df['month']==10,'season'] = 'wet'
    df.loc[df['month']==11,'season'] = 'wet'
    df.loc[df['month']==12,'season'] = 'wet'
```

```python
month_season_detector(dprvi_f)
```

```python
rain = pd.read_csv('D:/thesis_data/pluviometric_data/pluviometric_data_PNB_paper_01.csv', sep=',')
```

```python
month_season_detector(rain)
```

```python
fig, ax = plt.subplots(figsize=(20, 2.5))

ax.vlines(index[9], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[21], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[39], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[50], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[67], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[80], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[98], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[111], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[128], ymin=0.5, ymax=1.5, colors='k', linewidth=2)
ax.vlines(index[141], ymin=0.5, ymax=1.5, colors='k', linewidth=2)

ax.plot(index, ff_fs['DpRVI'], 'red', linewidth=5, label='Forest-Grassland')
ax.plot(index, ff_fc['DpRVI'], 'gold', linewidth=5, label='Savanna-Grassland')
ax.plot(index, fs_fc['DpRVI'], 'royalblue', linewidth=5, label='Forest-Savanna')
ax.set_xticks(ticks=index)
ax.set_xticklabels(labels=dprvi_f['mes'], fontdict={'fontsize': 15, 'rotation': -45, 'horizontalalignment': 'center'})
ax.tick_params(axis='y', length=5, width=1.5, labelsize=20)
ax.tick_params(axis='x', which='both', bottom=True, top=False, labelbottom=True, length=8, width=2)
ax.set_xlim([-1, 149])
ax.set_ylim([0.7, 1.05])
ax.set_ylabel('Overlap [%]', fontsize=20)
ax.locator_params(axis='x', nbins=60)
ax.locator_params(axis='y', nbins=5)

sec_0 = ax.twinx()
sec_0.bar(rain['date'], rain['gauge_mm'], alpha=0.5)
sec_0.tick_params(axis='y', length=5, width=1.5, labelsize=20)
sec_0.locator_params(axis='y', nbins=5)
sec_0.set_ylim([0, 80])
sec_0.set_ylabel('rain gauge [mm]', fontsize=15)

#ax.legend()
ax.grid()

plt.tight_layout()

# plt.savefig('C:/Users/jales/Desktop/dprvi_ovl.svg', dpi=None, facecolor='w', edgecolor='w',
#     orientation='landscape', format='svg', transparent=False)
```
