```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

import scipy.stats as stats
```

```python
ff_fs_ovl = pd.read_csv('D:/thesis_data/VEG_INDICES/ovl/FF_FS_OVL_500_bins.csv')
ff_fc_ovl = pd.read_csv('D:/thesis_data/VEG_INDICES/ovl/FF_FC_OVL_500_bins.csv')
fs_fc_ovl = pd.read_csv('D:/thesis_data/VEG_INDICES/ovl/FS_FC_OVL_500_bins.csv')
```

```python
ff_fs_jm = pd.read_csv('D:/thesis_data/VEG_INDICES/jm_dists/FF_FS_jm_dists.csv')
ff_fc_jm = pd.read_csv('D:/thesis_data/VEG_INDICES/jm_dists/FF_FC_jm_dists.csv')
fs_fc_jm = pd.read_csv('D:/thesis_data/VEG_INDICES/jm_dists/FS_FC_jm_dists.csv')
```

# DpRVI

```python
fig, ax = plt.subplots(1,3,figsize=(15,5))

lin_reg_ff_fs = stats.linregress(ff_fs_ovl['DpRVI'], ff_fs_jm['DpRVI'])
lin_reg_ff_fc = stats.linregress(ff_fc_ovl['DpRVI'], ff_fc_jm['DpRVI'])
lin_reg_fs_fc = stats.linregress(fs_fc_ovl['DpRVI'], fs_fc_jm['DpRVI'])

ax[0].scatter(ff_fs_ovl['DpRVI'], ff_fs_jm['DpRVI'])
ax[0].set_ylabel('Jeffries-Matusita Distance')
ax[0].set_xlim(0.86,1)
ax[0].set_ylim(0.70,0.75)
ax[0].annotate(f'y = {round(lin_reg_ff_fs.intercept,2)}{round(lin_reg_ff_fs.slope,2)}x',(0.94,0.746))
ax[0].annotate(f'r = {round(lin_reg_ff_fs.rvalue,2)}',(0.94,0.744))
ax[0].annotate(f'p_value = 5.56e-83',(0.94,0.742))
ax[0].set_title('Forest-Savanna')

ax[1].scatter(ff_fc_ovl['DpRVI'], ff_fc_jm['DpRVI'])
ax[1].set_xlabel('Overlap Coefficient')
ax[1].set_xlim(0.70,1)
ax[1].set_ylim(0.70,0.82)
ax[1].annotate(f'y = {round(lin_reg_ff_fc.intercept,2)}{round(lin_reg_ff_fc.slope,2)}x',(0.88,0.80))
ax[1].annotate(f'r = {round(lin_reg_ff_fc.rvalue,2)}',(0.88,0.795))
ax[1].annotate(f'p_value = 2.84e-93',(0.88,0.790))
ax[1].set_title('Forest-Grassland')

ax[2].scatter(fs_fc_ovl['DpRVI'], fs_fc_jm['DpRVI'])
ax[2].set_xlim(0.84,1)
ax[2].set_ylim(0.70,0.76)
ax[2].annotate(f'y = {round(lin_reg_fs_fc.intercept,2)}{round(lin_reg_fs_fc.slope,2)}x',(0.93,0.75))
ax[2].annotate(f'r = {round(lin_reg_fs_fc.rvalue,2)}',(0.93,0.747))
ax[2].annotate(f'p_value = 6.02e-78',(0.93,0.744))
ax[2].set_title('Savanna-Grassland')
```
